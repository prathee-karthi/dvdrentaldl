Distribution of Rental Rates (Histogram)


```python

# Set up the visual style
sns.set(style="whitegrid")

# Plot 1: Distribution of rental rates
plt.figure(figsize=(8, 5))
sns.histplot(df['rental_rate'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Rental Rates')
plt.xlabel('Rental Rate ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
#The histogram shows a tri-modal distribution of rental rates, with three distinct price points at $1.0, $3.0, and $5.0. 
# The most common rate is $1.0, suggesting a large number of budget rentals, while the mid-range and premium rates are also popular. 
# There are clear gaps between the price tiers, indicating a structured pricing strategy with no intermediate rates between $1.5-$2.5 and $3.5-$4.5. 
# This segmentation likely targets different customer segments or film types, with potential opportunities to explore mid-range pricing.
```

Bar Plot: Film Ratings Distribution


```python

# Assuming df is your DataFrame
plt.figure(figsize=(8, 5))

# Plot countplot
ax = sns.countplot(data=df, x='rating', palette='Set2')

# Set titles and labels
plt.title('Film Ratings Distribution')
plt.xlabel('Film Rating')
plt.ylabel('Count')

# Create a dictionary for the rating counts (from your data)
rating_counts = {'PG-13': 3203, 'NC-17': 2964, 'PG': 2903, 'R': 2868, 'G': 2475}

# Manually create a custom legend
handles = [plt.Line2D([0], [0], color=color, lw=6) for color in sns.color_palette('Set2', len(rating_counts))]
labels = [f"{rating}: {count}" for rating, count in rating_counts.items()]

# Add the legend to the plot
plt.legend(handles, labels, title='Rating Counts', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

```python
#The chart shows a balanced distribution of film ratings, with PG-13 being the most common (3203 films), followed by NC-17 (2964), PG (2903), and R (2868) ratings. 
# G-rated films are the least common (2475). 
# The data suggests that content for teens and older audiences (PG-13) is in high demand, while general audience (G) films are less frequent.
```

Scatter Plot: Rental Rate vs Amount Paid


```python
# Plot 3: Rental rate vs. amount paid
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='rental_rate', y='amount', hue='category', palette='Set1', alpha=0.7)
plt.title('Rental Rate vs Amount Paid by Category')
plt.xlabel('Rental Rate ($)')
plt.ylabel('Amount Paid ($)')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#The scatter plot illustrates the relationship between Rental Rate and Amount Paid across various Categories. 
# Rental rates are fixed at $1.0, $3.0, and $5.0, with amounts paid ranging from $2 to over $10. 
# This indicates that while different categories (e.g., Horror, Documentary) are available at each rental rate, factors beyond the rental price, such as fees or additional rentals, influence the total amount paid. 
# Overall, it reflects a structured pricing strategy with diverse offerings across genres.
```

Bar Plot: Category Distribution


```python
# Plot 4: Distribution of movie categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='category', palette='Set3', order=df['category'].value_counts().index)
plt.title('Movie Categories Distribution')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

```python
#Sports, Animation, and Action are the most popular movie categories.
#Music, Travel, and Horror are the least popular movie categories.
#There is a diverse range of genres to cater to different audience preferences.
#Genre popularity can help understand audience preferences and inform content creation and marketing strategies.
#Analyzing genre trends and box office performance can provide valuable insights for the film industry.
#Understanding genre preferences by demographic can help tailor content and marketing efforts.
```
